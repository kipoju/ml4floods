## Step 0: load the manually drawn polygons

Polygons could be given or manually drawn using http://geojson.io/, they could also be drawn using QGIS or an automatic script. If you draw them keep them small (250 x 250 km²)

```python
data = gpd.read_file("/home/kikeportales/Documentos/Investigacio/Floods/activations/pakistan/autoaois_pak.geojson")
data
```

```python
names_aois = []
for x in range(len(data.name)):
    if x <9:
        name = f'AOI0{x+1}'
    else:
        name = f'AOI{x+1}'
    names_aois.append(name)

```

```python

data['name'] = names_aois
data
```

## Step 2: Add the pickle files to the GCP location

```python
import datetime

for row in data.itertuples():
    ems_code = "EMSR586"
    aoi_code = row.name
    satellite_date = datetime.datetime(2022, 7, 5)
    data_save = {'satellite date': satellite_date,
                 'area_of_interest_polygon': row.geometry,
                 'ems_code': ems_code,
                 'aoi_code': aoi_code,
                 'date_ems_code': satellite_date}
    gs_path = f"gs://ml4cc_data_lake/0_DEV/1_Staging/operational/EMSR586/{aoi_code}/flood_meta/{satellite_date.strftime('%Y-%m-%d')}.pickle"
    utils.write_pickle_to_gcp(gs_path, data_save)
```

```python
import datetime
from ml4floods.data.utils import get_filesystem
fs = get_filesystem("gs://ml4cc_data_lake/0_DEV/1_Staging/operational/EMSR629/aois.json")

for row in data.itertuples():
    ems_code = "EMSR629"
    aoi_code = row.name
    satellite_date = datetime.datetime(2022, 8, 25)
    data_save = {'satellite date': satellite_date,
                 'area_of_interest_polygon': row.geometry,
                 'ems_code': ems_code,
                 'aoi_code': aoi_code,
                 'date_ems_code': satellite_date}
    gs_path = f"gs://ml4cc_data_lake/0_DEV/1_Staging/operational/EMSR629/{aoi_code}/flood_meta/{satellite_date.strftime('%Y-%m-%d')}.pickle"
    if not fs.exists(gs_path):
        print('No exists, writing', gs_path)
        utils.write_pickle_to_gcp(gs_path, data_save)
```
